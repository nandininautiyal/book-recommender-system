```python
num_rating=ratings_and_books.groupby('title')['rating'].count().reset_index()
```

```python
final.drop_duplicates(['user_id','title'],inplace=True)
```

```python
book_pivot=final.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse= csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    
    recommended_books = []
    
    for i in range(len(suggestion)):
        books_list = book_pivot.index[suggestion[i]]
        for j in books_list:
            img_url = books[books['title'] == j]['img_url'].values[0]  # Use 'books' instead of 'books_df'
            recommended_books.append((j, img_url))
    
    # Display book names and image URLs
    for book, img in recommended_books:
        print(f"Book: {book}")
        print(f"Image URL: {img}")

# Test the function
book_name = 'A Bend in the Road'
recommend_book(book_name)

```

```python
book_name='Harry Potter and the Chamber of Secrets (Book 2)'
recommend_book(book_name)
```

```python
# Merge books and ratings dataset
book_ratings = ratings.merge(books, on='ISBN')

# Compute book popularity based on the number of ratings
top_books = book_ratings.groupby(['title', 'img_url']).size().reset_index(name='num_ratings')

# Get the top 50 books with the most ratings
top_50_books = top_books.sort_values(by='num_ratings', ascending=False).head(50)

# Display the top 50 books with image links
top_50_books

```
